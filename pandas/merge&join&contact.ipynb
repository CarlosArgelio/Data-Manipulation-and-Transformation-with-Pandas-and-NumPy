# Merge

```python
izq = pd.DataFrame({
    'key': ['key0', 'key1', 'key2', 'key3'],
    'A': ['A0', 'A1','A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

der = pd.DataFrame({
    'key': ['key0', 'key1', 'key2', 'key3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3'],
})
```

```python
izq
```

```python
der
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({
    'key': ['key0', 'key1', 'key2', 'key3'],
    'A': ['A0', 'A1','A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

der = pd.DataFrame({
    'key_2': ['key0', 'key1', 'key2', 'key3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3'],
})
```

```python
izq.merge(der, on='key')
```

```python
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq = pd.DataFrame({
    'key': ['key0', 'key1', 'key2', 'key3'],
    'A': ['A0', 'A1','A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

der = pd.DataFrame({
    'key_2': ['key0', 'key1', 'key2', np.nan],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3'],
})
```

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

# Join - index match

```python
izq = pd.DataFrame({
    'A': ['A0', 'A1','A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['k0', 'k1', 'k2'])

der = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2'],
}, index=['k0', 'k1', 'k2'])
```

```python
izq.join(der, how='ineer')
```
